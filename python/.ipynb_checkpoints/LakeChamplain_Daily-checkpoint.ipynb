```python
W = '04279085' # WHITEHALL
R = '04295000' # RICHELIEU

start, stop = '1998-10-01','2017-06-14' 

Wdata = 'https://waterdata.usgs.gov/nwis/dv?cb_62614=on&format=rdb&site_no={}&referred_module=sw&period=&begin_date={}&end_date={}'.format(W, start, stop)
Rdata = 'https://waterdata.usgs.gov/nwis/dv?cb_62614=on&format=rdb&site_no={}&referred_module=sw&period=&begin_date={}&end_date={}'.format(R, start, stop)
```

```python
dfw = pd.read_csv(Wdata, skiprows=29, sep='\t') 
dfw.rename(columns = {'105873_62614_00003':'Stage'}, inplace=True)
dfw.drop(0, axis=0, inplace=True)
dfw['datetime'] = pd.to_datetime(dfw['datetime'], format= '%Y-%m-%d')
dfw['Stage'] = pd.to_numeric(dfw['Stage'])
dfw.head(2)
```

## Whitehall


```python
x = dfw['datetime']
y = dfw['Stage']

fig, ax1 = plt.subplots()
ax1.plot(x, y, 'b')
ax1.set_ylabel('', color='b')
ax1.set_xlabel('Time')
ax1.grid()
```

```python
#--Identify Peaks
dfw_peaks = dfw.query('Stage > 100').copy()
dfw_peaks.head()

#--manually eliminate neighbors
print(len(dfw_peaks))
for i in dfw_peaks.index:
    if i + 1 in dfw_peaks.index:
        dfw_peaks.drop(i+1, axis=0, inplace=True)
print(len(dfw_peaks))

dfw_idx = dfw_peaks.index
```

```python
dfr = pd.read_csv(Rdata, skiprows=29, sep='\t') 
dfr.rename(columns = {'105881_62614_00003':'Stage'}, inplace=True)
dfr.drop(0, axis=0, inplace=True)
dfr['datetime'] = pd.to_datetime(dfr['datetime'], format= '%Y-%m-%d')
dfr['Stage'] = pd.to_numeric(dfr['Stage'])
dfr.head(2)
```

```python
#--Identify Peaks
dfr_peaks = dfr.query('Stage > 99').copy()
dfr_peaks.head()

#--manually eliminate neighbors
print(len(dfr_peaks))
for i in dfr_peaks.index:
    if i + 1 in dfr_peaks.index:
        dfr_peaks.drop(i+1, axis=0, inplace=True)
print(len(dfr_peaks))

dfr_idx = dfr_peaks.index
```

```python
x = dfr['datetime']
y = dfr['Stage']

fig, ax1 = plt.subplots()
ax1.plot(x, y, 'b')
ax1.set_ylabel('', color='b')
ax1.set_xlabel('Time')
ax1.grid()
```

```python
idx = [554, 929, 1913, 2400, 2607, 2794, 3121, 3478, 4205, 4576, 5387, 5675]

def plotter(dfw, dfr, t):
    start, stop = t-60,  t+100
    x = dfw.datetime.ix[start:stop]
    y = dfw.Stage.ix[start:stop]

    x1 = dfr.datetime.ix[start:stop]
    y1 = dfr.Stage.ix[start:stop]

    plt.plot(x,y)
    plt.plot(x1,y1)
    plt.grid()
    plt.legend(['Whitehall (US)', 'Richeleiu (DS)'])
```

```python
plotter(dfw, dfr, idx[0])
```
